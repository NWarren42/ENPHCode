```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

aluminumSusceptibility = 1.65e-5
copperSusceptibility = -5.46e-6

numPoints = 1000

A = 1
mu_0 = 4*np.pi*1e-7
g = 9.81

aluminumSlope = aluminumSusceptibility*A/(2*mu_0*g) 
copperSlope = copperSusceptibility*A/(2*mu_0*g)

fieldValuesSquared_T2 = np.linspace(-1, 1, numPoints)
aluminumMassChanges_kg = aluminumSlope*fieldValuesSquared_T2
copperMassChanges_kg = copperSlope*fieldValuesSquared_T2

fig, ax = plt.subplots(figsize=(10, 6))

ax.axhline(0, color='black', lw=0.5)
ax.axvline(0, color='black', lw=0.5)

ax.plot(fieldValuesSquared_T2, aluminumMassChanges_kg, 'b-', label='Aluminum mass change')
ax.plot(fieldValuesSquared_T2, copperMassChanges_kg, 'r-', label='Copper mass change')

ax.set_xlabel(r'$Field \ Strength \ Squared \ [T^2]$')
ax.set_ylabel(r'$Mass \ Change \ [kg] $')

ax.text(0.6, 0.25, r'$\chi_{Al} = 1.65 \times 10^{-5}$' + 
        f'\n $slope = {aluminumSlope:.2f} A$'  + r'$\frac{m_0}{H_0^2}$')
ax.text(-0.7, 0.2, r'$\chi_{Cu} = -5.46 \times 10^{-6}$'  + 
        f'\n $slope = {copperSlope:.2f} A$' + r'$\frac{m_0}{H_0^2}$')

ax.set_xticks([-1, -0.5, 0, 0.5, 1])
ax.set_xticklabels([r"$- H_0^2$", r"$- \frac{H_0^2}{2}$", r"$0$", r"$ \frac{H_0^2}{2}$", r"$H_0^2$"])

ax.set_yticks([-max(aluminumMassChanges_kg), -0.5*max(aluminumMassChanges_kg), 0, 0.5*max(aluminumMassChanges_kg), max(aluminumMassChanges_kg)])
ax.set_yticklabels([r"$- m_0$", r"$- \frac{m_0}{2}$", r"$0$", r"$ \frac{m_0}{2}$", r"$m_0$"])

ax.set_xlim(-1, 1)
ax.set_ylim(-max(aluminumMassChanges_kg), max(aluminumMassChanges_kg))

ax.legend()

plt.show()
```
